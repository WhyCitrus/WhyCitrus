# AutoEncoder & MLP


```python
train_df=pd.read_csv("../../TfProjects/Torch Module/train.csv", index_col='id')
test_df=pd.read_csv("../../TfProjects/Torch Module/test.csv", index_col='id')
```

```python
'''get training feat'''

non_float_columns = train_df.select_dtypes(exclude=['float']).columns
categorical_features = list(non_float_columns)
categorical_features.remove('NObeyesdad')

float_columns = train_df.select_dtypes(include=['float']).columns
numerical_features = list(float_columns)

target = 'NObeyesdad'
```

```python
from sklearn.preprocessing import LabelEncoder

class ExtendedLabelEncoder(LabelEncoder):
    def transform(self, y):
        unseen_labels = set(y) - set(self.classes_)
        if unseen_labels:
            # 动态扩展 classes_
            self.classes_ = np.append(self.classes_, list(unseen_labels))
        return super().transform(y)

# 初始化扩展版 LabelEncoder
label_encoder = ExtendedLabelEncoder()
encoding_mapping = {}

for column in non_float_columns:
    # 对训练集进行 fit_transform
    train_df[column] = label_encoder.fit_transform(train_df[column])
    encoding_mapping[column] = dict(enumerate(label_encoder.classes_))
    if column != 'NObeyesdad':
        # 对测试集进行 transform
        test_df[column] = label_encoder.transform(test_df[column])

print("训练集编码完成:")
print(train_df)
print("测试集编码完成:")
print(test_df)
```

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Concatenate, Dropout, BatchNormalization, LeakyReLU

# 输入层
input_layer = Input(shape=(16,), name="Input")

# Encoder 部分
encoded = Dense(64, name="Encoder_Layer1")(input_layer)
encoded = LeakyReLU(alpha=0.2)(encoded)
encoded = BatchNormalization(name="Encoder_BatchNorm1")(encoded)
encoded = Dropout(0.2, name="Encoder_Dropout1")(encoded)

encoded = Dense(32, name="Encoder_Layer2")(encoded)
encoded = LeakyReLU(alpha=0.3)(encoded)
encoded = Dropout(0.3, name="Encoder_Dropout2")(encoded)

# Decoder 部分
decoded = Dense(64, activation='relu', name="Decoder_Layer1")(encoded)
decoded_output = Dense(16, activation='linear', name="Decoder_Output")(decoded)

# Autoencoder 的分类任务
autoencoder_classification = Dense(7, activation='softmax', name="Autoencoder_Classification")(encoded)

# MLP 部分
mlp_input = Concatenate(name="MLP_Input")([input_layer, encoded])
mlp = Dense(64, name="MLP_Layer1")(mlp_input)
mlp = LeakyReLU(alpha=0.2)(mlp)
mlp = BatchNormalization(name="MLP_BatchNorm1")(mlp)
mlp = Dropout(0.3, name="MLP_Dropout1")(mlp)

mlp = Dense(32, name="MLP_Layer2")(mlp)
mlp = Dropout(0.3, name="MLP_Dropout2")(mlp)

final_output = Dense(7, activation='softmax', name="Final_Output")(mlp)

# 定义模型
model_plus = Model(inputs=input_layer, outputs=[autoencoder_classification, final_output, decoded_output])

# 编译模型
model_plus.compile(
    optimizer='adam',
    loss={
        "Autoencoder_Classification": "sparse_categorical_crossentropy",
        "Final_Output": "sparse_categorical_crossentropy",
        "Decoder_Output": "mse"
    },
    loss_weights={
        "Autoencoder_Classification": 0.8,
        "Final_Output": 1.5,
        "Decoder_Output": 0.8
    },
    metrics={
        "Autoencoder_Classification": "accuracy",
        "Final_Output": "accuracy"
    }
)
```

```python
# 输入层
input_layer = Input(shape=(16,), name="Input")

# Encoder 部分
encoded = Dense(64, activation='relu', name="Encoder_Layer1")(input_layer)
encoded = Dense(32, activation='relu', name="Encoder_Layer2")(encoded)

# Decoder 部分
decoded = Dense(64, activation='relu', name="Decoder_Layer1")(encoded)
decoded_output = Dense(16, activation='linear', name="Decoder_Output")(decoded)

# Autoencoder 的分类任务
autoencoder_classification = Dense(7, activation='softmax', name="Autoencoder_Classification")(encoded)

# MLP 分类任务
mlp_input = Concatenate(name="MLP_Input")([input_layer, encoded])
mlp = Dense(64, activation='relu', name="MLP_Layer1")(mlp_input)
mlp = Dense(32, activation='relu', name="MLP_Layer2")(mlp)
final_output = Dense(7, activation='softmax', name="Final_Output")(mlp)

# 定义模型
model = Model(inputs=input_layer, outputs=[autoencoder_classification, final_output, decoded_output])

# 编译模型
model.compile(
    optimizer='adam',
    loss={
        "Autoencoder_Classification": "sparse_categorical_crossentropy",  # 使用稀疏分类交叉熵
        "Final_Output": "sparse_categorical_crossentropy",  # 使用稀疏分类交叉熵
        "Decoder_Output": "mse"  # 自编码器的重构损失
    },
    loss_weights={
        "Autoencoder_Classification": 0.8,
        "Final_Output": 1.0,
        "Decoder_Output": 0.8
    },
    metrics={
        "Autoencoder_Classification": "accuracy",
        "Final_Output": "accuracy"
    }
)
```
